### Introduction
In this workbook we will try to understand concepts behind plotting data in matplotlib. Matplotlib is probably the most common library used for data visualization in Python. It is used extensively and is a pretty flexible library. 
<br>We will try to understand the concepts using basic plots first and then slowly move to real-life data sets. 
<br>Let us understand the structure of the matplotlib first. We generally import pyplot from matplotlib which is the plot generating function of matplotlib library. 
<br>Pyplot acts as an altering interface and every change in pyplot gets stacked; when we call plt.show() all those changes are reflected in form of a chart. 
<br>Let us do all of this via an example

By now it should be clear how to create a basic plot. There are multiple elements to a graph:
1. Figure - Is the soul of the graph which we will be drawing
2. Axes - They contain elements like Titles, Legends, Grids etc.
3. Title - Title of the graph
4. Legend - Legends required
5. Ticks - Major and minor ticks of the data
6. Labels - Label certain elements of the graph,
7. Grids - Horizontal and Vertical lines in the background
8. Lines/ Markers - Actual markers - dots, lines etc which are used to represent the data

Matplotlib doesn't support centimeters as a unit as of now. However, we can write a conversion function easily from inches to cms. 

````python
import numpy as np
def cm_to_inch(value):
    return value/2.54
#This function will return values in inches when you input in cms and then can be used to create a plot with defined cms. 
plt.figure(figsize = (cm_to_inch(15),cm_to_inch(10)))
x = np.arange(0,10,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
````

```fig = plt.figure()```. Now fig can have multiple types of elements which can be added to. Example - ```set_figheight, set_figwidth or add_subplot```. 
These come handy when we want to add multiple elements to the figure in series. Remember everytime you add an element it gets added to the existing figure and is then displayed together. 

````python
##Let us take an example
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(10)
ax = fig.add_subplot(121) #NROWS = 1, NCOLS = 2 and INDEX = 1
ax2 = fig.add_subplot(122)
ax.plot(x,y) #AX and AX2 becomes the plot altering features once we have added the subplots.
ax.set_xlabel("X VALUES")
ax.set_ylabel("SIN X")
ax2.plot(x,z)
plt.show()
````

Adding text to the plot becomes an important way to tell people what you are trying to communicate. There are multiple methods which can be used to add text to your plots:
1. plt.text() or ax.text()
2. plt.annotate() or ax.annotate()
3. plt.xlabel() or ax.set_xlabel() - same for ylabel also
4. plt.title() or ax.set_title()
5. plt.suptitle() or fig.suptitle()
<br>Let us run a blank graph with all the elements above. 

````python
fig,ax = plt.subplots()
fig.suptitle("This is a figure level suptitle")
ax.set_xlabel("X LABEL")
ax.set_ylabel("Y LABEL")
ax.text(0.5,0.5, "This is a generic text")
ax.annotate("This is an annotation between \n generic text and itself",xy = (0.625,0.5),xytext = (0.25,0.25),arrowprops = dict
           (arrowstyle = "<->",connectionstyle = "arc3, rad = 0.15"))
plt.show()
````

Annotate has a lot of use in explaining your data well and has a lot of options. Let us explore the same using some code. 

````python
fig,ax = plt.subplots()
fig.suptitle("Different type of annotations")
ax.annotate("Typical Annotation",
            xy = (0.8,0.1),
            xytext = (0.1,0.1),
            arrowprops=dict(facecolor = 'black',width = 1,headwidth = 10),
            verticalalignment = 'center')
ax.annotate("With arrow style my style",xy = (0.8,0.2)
            ,xytext = (0.1,0.2),
           arrowprops = dict(arrowstyle = "->"),
            verticalalignment = 'center')
ax.annotate("Annotation with arrow style",
           xy = (0.8,0.3),
           xytext = (0.1,0.3),
           arrowprops=dict(arrowstyle="<->"),
                          verticalalignment = 'center')
ax.annotate("Annotation with arrow style",
           xy = (0.8,0.4),
           xytext = (0.1,0.4),
           arrowprops= dict(arrowstyle = "|-|"),
           verticalalignment = 'center')
ax.annotate("Annotation with arrow style",
           xy = (0.8,0.5),
           xytext=(0.1,0.5),
           arrowprops = dict(arrowstyle = "fancy"),
           verticalalignment = 'center')
plt.show()
````

````python
#Let us take one more example of annotations which can explain how exactly it can work in the real life example.
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(15,10))
x = [2,2,2,2]
y = [3,4,5,6]
plt.xlim(0,2)
plt.ylim(0,10)
plt.scatter(x,y)
ax.annotate("Connection style with angle",
           xy = (2,3),
           xytext = (0,4),
           arrowprops =dict(arrowstyle = "->",connectionstyle = "angle,angleA = 120,angleB= 0"),
           verticalalignment = 'center')
ax.annotate("Connection style with Angle3",
           xy = (2,4),
           xytext = (0,5),
           arrowprops = dict(arrowstyle = "->",connectionstyle = "angle3,angleA = 90,angleB = 0"),
           verticalalignment = 'center')
ax.annotate("Connection with arc",
           xy = (2,5),
           xytext = (0,6),
           arrowprops = dict(arrowstyle = "->",connectionstyle = "arc,angleA = 0,angleB = 0,armA=0,armB=45"),
           verticalalignment = 'center')
ax.annotate("Connection wtih Arc3",
           xy = (2,6),
           xytext = (0,6.5),
           arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=0.25"),
           horizontalalignment = 'left')
plt.show()
````

Legends are important features of the graph and are used for representing data better. There are multiple ways of adding legends.

````python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots()
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(x,y,color='blue',label = 'sine wave')
ax.plot(x,z,color='black',label = 'cosine wave')
leg = ax.legend()
plt.show()
````

We can change the positions of the legend and also try to change some parameters like frame. There are other parameters of legend as well which can be changed like - Font Size, Title (& Title Font Size), shadow etc. let us look at one simple and one more complex example.

````python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize = (12,6))
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(x,y,color='blue',label = 'sine wave')
ax.plot(x,z,color='black',label = 'cosine wave')
leg = ax.legend(loc='upper right',frameon =False)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize = (12,6))
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(x,y,color='blue',label = 'sine wave')
ax.plot(x,z,color='black',label = 'cosine wave')
leg = ax.legend(title = "Functions",fontsize = 12,
                title_fontsize = 14,
                loc = 'center',
                bbox_to_anchor=(0.5,0.5),
                shadow=True,ncol=2)
plt.show()
````

Font sizes are critical part of a graph and when you increase the size of the graph generally the font-sizes do not scale as per our needs thus we might need to control the font sizes. Let us see a few examples how to do the same.

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
fig,ax = plt.subplots(figsize = (15,10))
plt.plot(y,color = 'blue',label = 'sine wave')
plt.plot(z,color = 'red',label = 'cosine wave')
ax.set_title("Sine and Cosine Waves")
ax.set_xlabel("Time")
ax.set_ylabel("Intensity")
plt.show()
#Look at the fontsizes of x-title and y-title they are too small to be seen properly we can change the font size as followabs
````

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
fig,ax = plt.subplots(figsize = (15,10))
plt.plot(y,color = 'blue',label = 'sine wave')
plt.plot(z,color = 'red',label = 'cosine wave')
ax.set_title("Sine and Cosine Waves",fontsize = 20)
ax.set_xlabel("Time",fontsize = 16)
ax.set_ylabel("Intensity",fontsize = 16)
plt.show()
#This looks a lot better !! 
````

You can save the figures in your desired locations using ```plt.savefig('figure_name`)``` command this will save a png file. 

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
fig,ax = plt.subplots(figsize = (15,10))
plt.plot(y,color = 'blue',label = 'sine wave')
plt.plot(z,color = 'red',label = 'cosine wave')
ax.set_title("Sine and Cosine Waves",fontsize = 20)
ax.set_xlabel("Time",fontsize = 16)
ax.set_ylabel("Intensity",fontsize = 16)
plt.savefig('test_fig')
#You will see your figure is saved in the location of this notebook with name test_fig.png
````

We can also define the resolution of the saved figure using ```plt.savefig("Name of the File", dpi = XYZ)``` higher the value of XYZ higher will be the resolution. 
We can also use parameter transparent which will set the transparency of the graph. 
<br>Remember increasing the DPI increases the image size example for 1000 dpi we may need close to 1.26 MB whereas for 100 dpi the size required will be 73 KB.

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
fig,ax = plt.subplots(figsize = (15,10))
plt.plot(y,color = 'blue',label = 'sine wave')
plt.plot(z,color = 'red',label = 'cosine wave')
ax.set_title("Sine and Cosine Waves",fontsize = 20)
ax.set_xlabel("Time",fontsize = 16)
ax.set_ylabel("Intensity",fontsize = 16)
plt.savefig('test_fig50',dpi = 50)
plt.savefig('test_fig100',dpi = 100)
plt.savefig('test_fig1000',dpi = 1000)
plt.savefig('test_fig1000',dpi = 1000,transparent = True)
plt.savefig('test_fig1000',dpi = 1000,transparent = True,facecolor = 'red')


#You will see your figure is saved in the location of this notebook with name test_fig.png
````

#### Playing with AXES:
We can play with axes to generate many effects which can be used for creating multiple type of graphs.
1. X-limits and Y-Limits
2. Ticks of axes - where we can change the #Ticks and #Tick Angles

````python
##Example of ticks
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize = (12,6))
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(x,color='blue',label = 'sine wave')
ax.plot(z,color='black',label = 'cosine wave')

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize = (12,6))
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax.plot(y,color='blue',label = 'Sine wave')
ax.plot(z, color='green',label = 'Cosine Wave')
plt.xlim([25,50])
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize = (15,8))
x = np.arange(0,10,0.1)
y = np.sin(x)
z = np.cos(x)
ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax.set_title("Full View")
ax.plot(y,color='blue',label='Sine Wave')
ax.plot(z,color='black',label = 'Cosine Wave')

ax2.set_title("Truncated View")
ax2.plot(y,color='blue',label='Sine Wave')
ax2.plot(z,color='black',label = 'Cosine Wave')

ax2.set_xlim([25,50])

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.subplots(figsize = (12,6))
x = np.random.randint(low = 0,high = 50, size = 100)
y = np.random.randint(low = 0,high = 50, size = 100)
plt.xticks(np.arange(0,len(x)+1,1))
plt.yticks(np.arange(0,len(y)+1,2))
plt.xlim([0,10])
plt.plot(x,color='blue')
plt.plot(y,color='black')
plt.show()
````
