### Optimization using differentiation

```python
import sympy as sp
import numpy as np
```

```python
x = sp.symbols('x')
pi = sp.symbols('pi')
```

```python
I = (np.pi*x*(10-x)**2)/3
I
```

```python
k = sp.solve(sp.Eq(sp.diff(I,x),0))
```

```python
I.subs(x,float(k[0]))
```

```python
a = np.arange(0,20,0.05)
```

```python
a
```

```python
max([I.subs(x,val) for val in a])
```
